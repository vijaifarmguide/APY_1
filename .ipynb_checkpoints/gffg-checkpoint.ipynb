```python
import pandas as pd
# turn of warning messages
pd.options.mode.chained_assignment = None  # default='warn'

# get data
df = pd.read_csv('APY_data_simple crops.csv')
#df.head(5)


#remove numbers
```

```python
def describe_dataframe(df=pd.DataFrame()):
    """This function generates descriptive stats of a dataframe
    Args:
        df (dataframe): the dataframe to be analyzed
    Returns:
        None

    """
    print("\n\n")
    print("*"*30)
    print("About the Data")
    print("*"*30)
    
    print("Number of rows::",df.shape[0])
    print("Number of columns::",df.shape[1])
    print("\n")
    
    print("Column Names::",df.columns.values.tolist())
    print("\n")
    
    print("Column Data Types::\n",df.dtypes)
    print("\n")
    
    print("Columns with Missing Values::",df.columns[df.isnull().any()].tolist())
    print("\n")
    
    print("Number of rows with Missing Values::",len(pd.isnull(df).any(1).nonzero()[0].tolist()))
    print("\n")
    
    print("Sample Indices with missing data::",pd.isnull(df).any(1).nonzero()[0].tolist()[0:5])
    print("\n")
    
    print("General Stats::")
    print(df.info())
    print("\n")
    
    print("Summary Stats::")
    print(df.describe())
    print("\n")
    
    print("Dataframe Sample Rows::")
    display(df.head(5))
```

```python
%time describe_dataframe(df)
```

```python
df1=df[df.District!=" "]
df1['District']=df1['District'].str.replace('[^a-zA-Z ]', '')
```

```python
#Yield =0 ie , Production = 0
df1[df1['Production (Tonnes)'].isnull()]=0
```

```python
df1=df1[df1['Season']!=0]
```

```python
yearlist=[' 2009-10', ' 2010-11', ' 2011-12', ' 2012-13', ' 2013-14', ' 2014-15', ' 2015-16']
for year in yearlist:
    df2=df1[(df1['Year']==year) & ((df1['Season']=='Rabi       ') | (df1['Season']=='Whole Year ')) & ((df1.State=="Bihar") | (df1.State=="Chhattisgarh")  | (df1.State=="Gujarat") | (df1.State=="Haryana") | (df1.State=="Madhya Pradesh") | (df1.State=="Maharashtra") | (df1.State=="Punjab") | (df1.State=="Uttar Pradesh") | (df1.State=="Rajasthan"))].sort_values(['State','District', 'Area (Hectare)'], ascending=[True,True, False])
    df3=df2[[ 'State', 'District', 'Crop', 'Season', 'Area (Hectare)' ]]
    df3['TotalDistrictArea'] =  df3.groupby(['State','District'],sort=False)['Area (Hectare)'].transform('sum')
    df3 = df3.groupby(['State','District'],sort=False).head(2).reset_index(drop=True)
    df3['TopTwoCropAreaPercent'] = df3['Area (Hectare)']/df3['TotalDistrictArea']
    df3.to_csv("toptwopercentRabi{}.csv".format(year.strip()))
```
